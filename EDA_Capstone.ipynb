```python
import pandas as pd
```

```python
file_path = '~/Downloads/Food_Inspection.csv'
df = pd.read_csv(file_path)
```

```python
print(df.info())
```

```python
null_summary = df.isnull().sum()
print("Null Value Summary for All Columns:")
print(null_summary)
```

```python
Category_distinct = df['Category'].unique()

# Display the distinct values
print(Category_distinct)
```

```python
df_new = df[~df['Category'].str.contains('Do Not Use', na=False)]
```

```python
Category_distinct = df_new['Category'].unique()

# Display the distinct values
print(Category_distinct)
```

```python
null_inspection_type_rows = df_new[df_new['Inspection_type'].isnull()]

# Display the rows with null 'inspection_type'
print(null_inspection_type_rows)
```

```python
df_cleaned = df_new.dropna(subset=['Inspection_type'])
```

```python
num_rows = df.shape[0]
print(f"Number of rows in the DataFrame: {num_rows}")

num_rows = df_cleaned.shape[0]
print(f"Number of rows in the DataFrame: {num_rows}")
```

```python
# Frequency Analysis of Restaurant Categories
restaurant_category_counts = df_cleaned['Category'].value_counts()
print("\nRestaurant Category Frequency:")
print(restaurant_category_counts)

# Inspection Results Overview
inspection_results_counts = df_cleaned['Inspection_results'].value_counts()
print("\nInspection Results Frequency:")
print(inspection_results_counts)

# City-wise Analysis
city_grouped = df_cleaned.groupby('City')
city_inspection_results_counts = city_grouped['Inspection_results'].value_counts()
print("\nCity-wise Inspection Results:")
print(city_inspection_results_counts)

# Inspection Type Comparison
inspection_type_counts = df_cleaned['Inspection_type'].value_counts()
print("\nInspection Type Frequency:")
print(inspection_type_counts)
```

```python
# Types of inspections and their frequency
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(8, 6))
sns.countplot(data=df_cleaned, x='Inspection_type')
plt.title("Types of Inspections and Their Frequency")
plt.xlabel("Inspection Type")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show()
```

```python
food_complaints_df = df_cleaned[df_cleaned['Inspection_type'] == "Food Complaint"]

# Group the filtered data by city and count the number of complaints in each city
city_complaint_counts = food_complaints_df['City'].value_counts()

# Get the top 10 cities with the most food complaints
top_10_cities = city_complaint_counts.head(10)

# Display the top 10 cities and their food complaint counts
print("Top 10 Cities with the Most Food Complaints:")
print(top_10_cities)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df_cleaned is your DataFrame
# Filter rows where Owner is 'MARK FOSSETT'
mark_fossett_df = df_cleaned[df_cleaned['Owner'] == 'MARK FOSSETT']

# List of columns to analyze
columns_of_interest = ['Food_from_approved_source', 'Food_protected_from_contamination', 
                        'Ill_workers_restricted', 'Proper_sewage_disposal', 'Rodent_and_insects']

# Count occurrences of 'In Compliance' and 'Out of Compliance' for each column
compliance_counts = {}
for column in columns_of_interest:
    compliance_counts[column] = mark_fossett_df[column].value_counts()

# Plotting the counts as bar charts
plt.figure(figsize=(12, 8))

for i, (column, counts) in enumerate(compliance_counts.items(), start=1):
    plt.subplot(2, 3, i)
    plt.bar(counts.index, counts.values, color=['green', 'red'])
    plt.title(f"{column} Status")
    plt.xlabel("Compliance Status")
    plt.ylabel("Count")
    plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
